```python
data = data.fillna(data['fare_amount'].median())
```

```python
data.isna().sum()
```

```python
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns
```

```python
for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.scatterplot(x=data.index, y=data[col], alpha=0.5) # alpha 
    plt.title(f'{col} Scatter Plot')
    plt.xlabel('Index')
    plt.ylabel(col)
    plt.show()
```

```python
col = 'passenger_count' 
```

```python
Q1 = data[col].quantile(0.25)
```

```python
Q3 = data[col].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
data_cleaned = data[(data[col] >= Q1 - 1.5 * IQR) & (data[col] <= Q3 + 1.5 * IQR)] 
```

```python
data_cleaned.shape
```

```python
data_cleaned['tpep_pickup_datetime'] = pd.to_datetime(data_cleaned['tpep_pickup_datetime'])
```

```python
data_cleaned['tpep_dropoff_datetime'] = pd.to_datetime(data_cleaned['tpep_dropoff_datetime'])
```

```python
data_cleaned['trip_duration'] = (data_cleaned['tpep_dropoff_datetime'] - data_cleaned['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
print(data_cleaned[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_duration']].head())
```

```python

plt.figure(figsize=(8, 5))
sns.scatterplot(x=data_cleaned['trip_duration'], y=data_cleaned['fare_amount'], alpha=0.5)
plt.title('Taxi Trip Duration vs Fare Amount')
plt.xlabel('Trip Duration (minutes)')
plt.ylabel('Fare Amount ($)')
plt.show()
```
